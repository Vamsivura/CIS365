### Importing the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Data set

```python
df = pd.read_csv("cardio_train.csv", sep=";")
df.head(5)
```

```python
df.info()
```

```python
df.describe() # descriptive stats of the dataset
```

### Plots describing the distribution of variables

```python
# Convert age from days to years
df['age_years'] = df['age'] // 365

# Plot histograms
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.hist(df['age_years'], bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(df['height'], bins=30, edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(df['weight'], bins=30, edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(df['ap_hi'], bins=30, edgecolor='black')
plt.title('Systolic Blood Pressure (ap_hi) Distribution')
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Finding the correlation between each variable

```python
# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
